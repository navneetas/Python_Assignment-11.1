```python
import pandas as pd
import numpy as np

titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv', sep=",")
titanic.head()
```

```python
titanic.sex.value_counts()
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

#Data for the plot
labels = 'Male', 'Female'
fracs = titanic.sex.value_counts()
colors = ['blue', 'red']

# Plot
plt.pie(fracs, colors=colors,
        autopct='%1.1f%%', shadow=False, startangle=140)

#To add labels, legend, title
plt.title("Male/Female Proportion") 
plt.axis('equal')
plt.legend(labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
plt.style.use('seaborn-white')

#Plot data
x = titanic['age']
y = titanic['fare']
color_mapping = {'male' : 'blue', 'female' : 'red'}   #To help display different colors as per gender.

# Scatterplot
titanic = titanic.dropna(subset=['sex'])
fig = plt.scatter(x, y, alpha=0.5, c=titanic['sex'].map(color_mapping))

#To add labels
plt.title("Scatterplot for Age and Fare")
plt.xlabel('Age')
plt.ylabel('Fare')

##To add legend
circ1 = Line2D([0], [0], linestyle="none", marker="o", alpha=0.4, markersize=10, markerfacecolor='blue')
circ2 = Line2D([0], [0], linestyle="none", marker="o", alpha=0.3, markersize=10, markerfacecolor='red')
plt.legend((circ1, circ2), ("Male", "Female"), numpoints=1, loc="best")

plt.show()
```
